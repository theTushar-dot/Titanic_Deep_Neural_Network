```python
def build_model(train_X_set, train_Y_set):
    
    X = train_X_set
    Y = train_Y_set
    m = len(train_X_set[0])
    
    alpha = 0.1
    
    #forward variable for layer 1 with 3 nodes
    nn_1 = 3 
    W_1 = np.random.randn(nn_1,4)
    b_1 = np.zeros((nn_1,1))
    Z_1 = 0
    A_1 = 0
    
    #forward variable for layer 2 with 1 nodes
    nn_2 = 1
    W_2 = np.random.randn(nn_2,nn_1)
    b_2 = 0
    Z_2 = 0
    A_2 = 0
    
    #backword varible of layer 2
    dZ_2 = 0
    dW_2 = 0
    db_2 = 0
    
    #backward varible of layer 1
    dZ_1 = 0
    dW_1 = 0
    db_1 = 0
    
    mH = 0
    
    for i in range(10000):
        
        #forward propogation
        
        Z_1 = np.dot(W_1,X) + b_1
        A_1 = np.maximum(0,Z_1)
        
        Z_2 = np.dot(W_2,A_1) + b_2
        A_2 = 1/(1+np.exp(-Z_2))
        
        Loss = -(1/m)*np.sum(Y*np.log(A_2)+((1-Y)*np.log(1-A_2)))
        
        # backword propogation
        
        dZ_2 = A_2 - Y
        
        dW_2 = (1/m)*np.dot(dZ_2,A_1.T)
        db_2 = (1/m)*np.sum(dZ_2,axis=1,keepdims=True)
        
        mH = Z_1
        
        mH[mH<=0] = 0
        mH[mH>0] = 1
        
        dZ_1 = np.dot(W_2.T,dZ_2)*mH
        
        dW_1 = (1/m)*np.dot(dZ_1,X.T)
        db_1 = (1/m)*np.sum(dZ_1,axis=1,keepdims=True)
        
        #updating
        
        W_1 = W_1 - (alpha*dW_1)
        W_2 = W_2 - (alpha*dW_2)
        
        b_1 = b_1 - (alpha*db_1)
        b_2 = b_2 - (alpha*db_2)
        
        print(i)
        print("--------------------------------------------------------")

        print("Loss",Loss)
        
    return W_1,W_2,b_1,b_2    
```

```python
W_1,W_2,b_1,b_2 = build_model(train_x,train_y)
```

```python
def predictor(test_x_set,w_1_set,w_2_set, b_1_set, b_2_set):
    X = test_x_set
    
    W_1 = w_1_set
    W_2 = w_2_set
    
    b_1 = b_1_set
    b_2 = b_2_set
    
    Z_1 = 0
    A_1 = 0
    
    Z_2 = 0
    A_2 = 0
   
    Z_1 = np.dot(W_1,X) + b_1
    A_1 = np.maximum(0,Z_1)
        
    Z_2 = np.dot(W_2,A_1) + b_2
    A_2 = 1/(1+np.exp(-Z_2))
    
    return A_2
```
